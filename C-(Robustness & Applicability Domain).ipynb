```python
# ─── Imports ──────────────────────────────────────────────────────────────────────────────
# 1. Random Seeds
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import shap
from scipy import stats

from rdkit import Chem, DataStructs
from rdkit.Chem import Descriptors, AllChem

from sklearn.model_selection import (
    train_test_split, StratifiedKFold,
    RandomizedSearchCV, cross_val_predict
)
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, matthews_corrcoef, confusion_matrix,
    classification_report, roc_curve, precision_recall_curve
)

from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, matthews_corrcoef
)
from sklearn.pipeline import Pipeline as SkPipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
# 2. Applicability Domain
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from rdkit import Chem, DataStructs
from rdkit.Chem import Descriptors, AllChem

from sklearn.model_selection import train_test_split, StratifiedKFold, RandomizedSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    roc_auc_score, matthews_corrcoef, confusion_matrix, roc_curve
)
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE

from sklearn.decomposition import PCA
from scipy.special import ndtr
from scipy.stats import chi2, norm
import shap
from scipy.special import ndtr



```

```python
# ─── FULL CLASSIFICATION WORKFLOW WITH MULTIPLE RANDOM SEEDS ─────

# 0) Define the seeds we'll run
SEEDS = [37, 42, 53, 69]

# 1) Load & prepare data once
df = (
    pd.read_csv(
        "M:/EPSRC Internship 2025/Project/Data/Classification Model/epimerisation_data_clean.csv",
        usecols=['epimerisation','smilesstarting_material']
    )
    .rename(columns={'smilesstarting_material':'smi_start'})
)
y_all = df['epimerisation'].astype(int).values

# Precompute RDKit descriptors
def compute_rdkit(smiles):
    names = [n for n,_ in Descriptors._descList]
    vals = []
    for smi in smiles:
        m = Chem.MolFromSmiles(smi)
        vals.append([f(m) if m else np.nan for _,f in Descriptors._descList])
    return pd.DataFrame(vals, columns=names).add_prefix("rdkit_")

rdkit_df = compute_rdkit(df['smi_start'])

# Precompute Morgan fingerprints
def compute_morgan(smiles, radius=2, n_bits=1024):
    fps, dens = [], []
    for smi in smiles:
        m = Chem.MolFromSmiles(smi)
        if m:
            fp = AllChem.GetMorganFingerprintAsBitVect(m, radius, nBits=n_bits)
            arr = np.zeros((n_bits,), dtype=int)
            DataStructs.ConvertToNumpyArray(fp, arr)
            fps.append(arr)
            dens.append(arr.sum()/n_bits)
        else:
            fps.append(np.full((n_bits,), np.nan))
            dens.append(np.nan)
    cols = [f"morgan_{i}" for i in range(n_bits)]
    df_bits = pd.DataFrame(fps, columns=cols)
    df_bits["morgan_density"] = dens
    return df_bits

morgan_df = compute_morgan(df['smi_start'])

# Combine and drop missing
X_all = pd.concat([rdkit_df, morgan_df], axis=1)
mask = X_all.notna().all(axis=1)
X_all = X_all[mask].reset_index(drop=True)
y_all = y_all[mask]

# make X and y available for the aggregate‐metrics cell below
X = X_all
y = y_all

# Container for overlay ROC curves
roc_overlay = {}

for SEED in SEEDS:
    print(f"\n\n================ RUN SEED={SEED} ================\n")
    random.seed(SEED)
    np.random.seed(SEED)

    # 2) Train/Test split
    X_train, X_test, y_train, y_test = train_test_split(
        X_all, y_all,
        test_size=0.2,
        random_state=SEED,
        stratify=y_all
    )

    # 3) Correlation filter transformer
    class CorrFilter(BaseEstimator, TransformerMixin):
        def __init__(self, thresh=0.9, feature_names=None):
            self.thresh = thresh
            self.feature_names = feature_names
        def fit(self, X, y=None):
            C = pd.DataFrame(X, columns=self.feature_names).corr().abs()
            U = C.where(np.triu(np.ones(C.shape), k=1).astype(bool))
            drop = [c for c in U.columns if any(U[c] > self.thresh)]
            self.keep = [i for i,f in enumerate(self.feature_names) if f not in drop]
            return self
        def transform(self, X): return X[:, self.keep]

    feat_names = X_train.columns.tolist()

    # 4) Build pipeline + hyperparameter tuning
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=SEED)
    pipe = ImbPipeline([
        ("imp", SimpleImputer(strategy="mean")),
        ("scale", StandardScaler()),
        ("corr", CorrFilter(thresh=0.9, feature_names=feat_names)),
        ("smote", SMOTE(random_state=SEED)),
        ("clf", RandomForestClassifier(
            class_weight="balanced", random_state=SEED
        ))
    ])
    param_dist = {
        "clf__n_estimators": [100, 200, 300],
        "clf__max_depth": [None, 10, 20],
        "clf__min_samples_split": [2, 5, 10]
    }
    rs = RandomizedSearchCV(
        pipe, param_distributions=param_dist, n_iter=20,
        scoring="precision", cv=cv, random_state=SEED, n_jobs=-1
    )
    rs.fit(X_train, y_train)
    best = rs.best_estimator_

    # 5) Test predictions
    y_test_pred  = best.predict(X_test)
    y_test_proba = best.predict_proba(X_test)[:,1]

    test_metrics = {
        "Accuracy": accuracy_score(y_test, y_test_pred),
        "Precision": precision_score(y_test, y_test_pred, zero_division=0),
        "Recall": recall_score(y_test, y_test_pred, zero_division=0),
        "F1": f1_score(y_test, y_test_pred, zero_division=0),
        "ROC AUC": roc_auc_score(y_test, y_test_proba),
        "MCC": matthews_corrcoef(y_test, y_test_pred)
    }
    cm = confusion_matrix(y_test, y_test_pred)

    # Print & table of test metrics
    print(">>> TEST SET METRICS (seed={})".format(SEED))
    display(pd.DataFrame(test_metrics, index=["Value"]).T.round(3))

    # Print classification report
    print(">>> Classification Report:\n")
    print(classification_report(y_test, y_test_pred, zero_division=0))

    # Plot confusion matrix
    plt.figure(figsize=(4,4))
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=["No","Yes"], yticklabels=["No","Yes"]
    )
    plt.title(f"Confusion Matrix (seed={SEED})")
    plt.xlabel("Predicted"); plt.ylabel("True")
    plt.tight_layout()
    plt.show()

    # 6) Y-randomization on test set
    rand_res = {m: [] for m in test_metrics}
    for _ in range(50):
        y_shuf = np.random.permutation(y_test)
        pred_r = best.predict(X_test)
        proba_r= best.predict_proba(X_test)[:,1]
        rand_res["Accuracy"].append(accuracy_score(y_shuf, pred_r))
        rand_res["Precision"].append(precision_score(y_shuf, pred_r, zero_division=0))
        rand_res["Recall"].append(recall_score(y_shuf, pred_r, zero_division=0))
        rand_res["F1"].append(f1_score(y_shuf, pred_r, zero_division=0))
        rand_res["ROC AUC"].append(roc_auc_score(y_shuf, proba_r))
        rand_res["MCC"].append(matthews_corrcoef(y_shuf, pred_r))

    rand_summary = []
    for m, vals in rand_res.items():
        arr = np.array(vals)
        mu, sd = arr.mean(), arr.std(ddof=1)
        observed = test_metrics[m]
        rand_mean = mu
        rand_std = sd
        z = (observed - rand_mean) / rand_std if rand_std > 0 else np.nan
        p = 1 - ndtr(z) if not np.isnan(z) else np.nan
        rand_summary.append({
            "Metric": m,
            "Real": round(observed,3),
            "Random Mean±SD": f"{mu:.3f}±{sd:.3f}",
            "Z": round(z,3),
            "p-value": round(p,3)
        })
    rand_df = pd.DataFrame(rand_summary).set_index("Metric")
    print(">>> Y-RANDOMIZATION (seed={})".format(SEED))
    display(rand_df)

    # 7) RF Feature importances
    keep_idx    = best.named_steps["corr"].keep
    final_feats = np.array(feat_names)[keep_idx]
    importances = best.named_steps["clf"].feature_importances_
    fi = pd.Series(importances, index=final_feats).sort_values(ascending=False)

    plt.figure(figsize=(8,6))
    fi.head(20).plot(kind="barh")
    plt.gca().invert_yaxis()
    plt.title(f"Top 20 RF Importances (seed={SEED})")
    plt.tight_layout()
    plt.show()

    # 8) ROC curve (per seed)
    fpr, tpr, _ = roc_curve(y_test, y_test_proba)
    roc_overlay[SEED] = (fpr, tpr, test_metrics["ROC AUC"])
    plt.figure(figsize=(6,4))
    plt.plot(fpr, tpr, label=f"AUC={test_metrics['ROC AUC']:.3f}")
    plt.plot([0,1],[0,1],'--',c='gray')
    plt.title(f"ROC Curve (seed={SEED})")
    plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend(); plt.tight_layout()
    plt.show()

    # 9) Precision–Recall curve (per seed)
    prec, rec, _ = precision_recall_curve(y_test, y_test_proba)
    ap = np.trapz(rec, prec)
    plt.figure(figsize=(6,4))
    plt.plot(rec, prec, label=f"AP={ap:.3f}")
    plt.title(f"Precision–Recall (seed={SEED})")
    plt.xlabel("Recall"); plt.ylabel("Precision"); plt.legend(); plt.tight_layout()
    plt.show()

# 10) Overlay ROC for all seeds
plt.figure(figsize=(7,5))
for seed,(fpr,tpr,auc) in roc_overlay.items():
    plt.plot(fpr, tpr, label=f"seed={seed} (AUC={auc:.3f})")
plt.plot([0,1],[0,1],'--',c='gray')
plt.title("Overlay Test ROC Curves (all seeds)")
plt.xlabel("False Positive Rate"); plt.ylabel("True Positive Rate")
plt.legend(loc="lower right"); plt.tight_layout()
plt.show()

print("\n=== WORKFLOW COMPLETE ===")
```

```python
# ── Applicability Domain on Training & Test Sets ──

def fit_applicability_domain(X_train, n_components=10, alpha=0.975, random_state=42):
    """
    Fit scaling + PCA on training data and determine Mahalanobis threshold.
    Returns scaler, pca, inverse covariance, mean, and chi2 threshold.
    """
    # Scale to zero mean/unit variance
    scaler = StandardScaler().fit(X_train)
    Xs = scaler.transform(X_train)
    
    # PCA reduction
    pca = PCA(n_components=n_components, random_state=random_state).fit(Xs)
    Xp = pca.transform(Xs)
    
    # Covariance in PCA space
    cov = np.cov(Xp, rowvar=False)
    inv_cov = np.linalg.inv(cov)
    mu = Xp.mean(axis=0)
    
    # Chi-squared threshold for given confidence level
    thresh = chi2.ppf(alpha, df=n_components)
    
    return scaler, pca, inv_cov, mu, thresh

def compute_applicability(X, scaler, pca, inv_cov, mu, thresh):
    """
    Compute Mahalanobis distance and in-domain flag for each sample in X.
    Returns distances and boolean in-domain mask.
    """
    Xs = scaler.transform(X)
    Xp = pca.transform(Xs)
    
    dists = []
    for vec in Xp:
        delta = vec - mu
        d2 = float(delta.dot(inv_cov).dot(delta))
        dists.append(d2)
    dists = np.array(dists)
    
    return dists, dists <= thresh

# Assuming X_train, X_test are already defined descriptor DataFrames

# 1) Fit AD on training set
scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh = fit_applicability_domain(
    X_train, n_components=10, alpha=0.975, random_state=42
)

print(f"Applicability Domain χ² threshold (97.5%): {chi2_thresh:.2f}")

# 2) Compute AD metrics on training set
train_dists, train_in_ad = compute_applicability(
    X_train, scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh
)
ad_train = pd.DataFrame({
    "Mahalanobis_D²": train_dists,
    "In_AD": train_in_ad
}, index=X_train.index)

# 3) Compute AD metrics on test set
test_dists, test_in_ad = compute_applicability(
    X_test, scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh
)
ad_test = pd.DataFrame({
    "Mahalanobis_D²": test_dists,
    "In_AD": test_in_ad
}, index=X_test.index)

# Display summary
print("\nTraining set AD coverage:")
print(ad_train["In_AD"].value_counts(), "\n")
print("Test set AD coverage:")
print(ad_test["In_AD"].value_counts())

# Optionally, append to your existing results tables:
# results_train = pd.concat([results_train, ad_train], axis=1)
# results_test  = pd.concat([results_test,  ad_test ], axis=1)
```

```python
# ─── MERGED WORKFLOW WITH APPLICABILITY DOMAIN & DUAL IMPORTANCES ─────

# 1) Load & prepare data (starting material + metadata)
df = pd.read_csv(
    "M:/EPSRC Internship 2025/Project/Data/Classification Model/epimerisation_data_clean.csv",
    usecols=[
        'epimerisation','smilesstarting_material',
        'conditions','reaction_class','substrate_class',
        'derivative_notes','specific_reagents',
        'solvent_1','solvent_2','temperature_C','reaction_time_h'
    ]
).rename(columns={'smilesstarting_material':'smi_start'})
y = df['epimerisation'].astype(int).values

# 2) Compute RDKit descriptors
def compute_rdkit(smiles):
    names = [n for n,_ in Descriptors._descList]
    rows = []
    for smi in smiles:
        m = Chem.MolFromSmiles(smi)
        rows.append([f(m) if m else np.nan for _,f in Descriptors._descList])
    return pd.DataFrame(rows, columns=names)
rdkit_df = compute_rdkit(df['smi_start'])

# 3) Compute 2048-bit Morgan fingerprints + density
def compute_morgan(smiles):
    fps, dens = [], []
    for smi in smiles:
        m = Chem.MolFromSmiles(smi)
        if m:
            fp = AllChem.GetMorganFingerprintAsBitVect(m,2,nBits=2048)
            arr = np.zeros((2048,),int)
            DataStructs.ConvertToNumpyArray(fp,arr)
            fps.append(arr); dens.append(arr.sum()/2048)
        else:
            fps.append([np.nan]*2048); dens.append(np.nan)
    cols = [f"FP_{i}" for i in range(2048)]
    df_fp = pd.DataFrame(fps, columns=cols)
    df_fp['FP_density'] = dens
    return df_fp
morgan_df = compute_morgan(df['smi_start'])

# 4) Process pre-reaction metadata
meta = df[[
    'conditions','reaction_class','substrate_class',
    'derivative_notes','specific_reagents',
    'solvent_1','solvent_2'
]].fillna('Missing')
meta_cat = pd.get_dummies(meta, drop_first=False)
meta_num = df[['temperature_C','reaction_time_h']].copy()
meta_num.fillna(meta_num.mean(), inplace=True)
meta_df = pd.concat([meta_cat, meta_num], axis=1)

# 5) Combine all features & drop any rows with missing values
X = pd.concat([rdkit_df, morgan_df, meta_df], axis=1)
mask = X.notna().all(axis=1)
X = X[mask].reset_index(drop=True)
y = y[mask]

# 5) Train/test split
SEED = 42
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=SEED
)

# 6) Correlation filter
class CorrFilter(BaseEstimator, TransformerMixin):
    def __init__(self, thresh=0.9, feature_names=None):
        self.thresh = thresh
        self.feature_names = feature_names
    def fit(self, X, y=None):
        C = pd.DataFrame(X, columns=self.feature_names).corr().abs()
        U = C.where(np.triu(np.ones(C.shape),1).astype(bool))
        drop = [c for c in U.columns if any(U[c]>self.thresh)]
        self.keep = [i for i,n in enumerate(self.feature_names) if n not in drop]
        return self
    def transform(self, X):
        return X[:,self.keep]

feat_names = X.columns.tolist()

# 7) Train Random Forest classifier with SMOTE
cv = StratifiedKFold(5, shuffle=True, random_state=SEED)
pipe = ImbPipeline([
    ("imp", SimpleImputer(strategy="mean")),
    ("scale", StandardScaler()),
    ("corr", CorrFilter(thresh=0.9, feature_names=feat_names)),
    ("smote", SMOTE(random_state=SEED)),
    ("clf", RandomForestClassifier(class_weight="balanced", random_state=SEED))
])
param_grid = {
    "clf__n_estimators": [100,200,300],
    "clf__max_depth": [None,10,20],
    "clf__min_samples_split": [2,5,10]
}
rs = RandomizedSearchCV(pipe, param_grid, n_iter=20,
                        scoring="precision", cv=cv,
                        random_state=SEED, n_jobs=-1)
rs.fit(X_train, y_train)
best = rs.best_estimator_

# 8) Applicability Domain via PCA-Mahalanobis
scaler_ad = StandardScaler().fit(X_train)
Xtr_s = scaler_ad.transform(X_train)
pca_ad = PCA(n_components=10, random_state=SEED).fit(Xtr_s)
Xtr_p = pca_ad.transform(Xtr_s)
cov = np.cov(Xtr_p, rowvar=False)
inv_cov = np.linalg.inv(cov)
mu = Xtr_p.mean(axis=0)
def mahal(x):
    d = x-mu
    return float(d.dot(inv_cov).dot(d))
threshold = chi2.ppf(0.975, df=Xtr_p.shape[1])
Xte_s = scaler_ad.transform(X_test)
Xte_p = pca_ad.transform(Xte_s)
md = np.array([mahal(x) for x in Xte_p])
in_dom = md <= threshold

# 9) Predict only in-domain
X_test_trans = best.named_steps["imp"].transform(X_test)
X_test_trans = best.named_steps["scale"].transform(X_test_trans)
X_test_trans = best.named_steps["corr"].transform(X_test_trans)
y_pred = best.named_steps["clf"].predict(X_test_trans)
y_proba = best.named_steps["clf"].predict_proba(X_test_trans)[:,1]

# mask out-of-domain
y_pred_dom = np.where(in_dom, y_pred, np.nan)
y_test_dom = np.where(in_dom, y_test,   np.nan)
y_proba_dom= np.where(in_dom, y_proba,  np.nan)

# 10) Compute metrics on in-domain
metrics = {
    "Accuracy": accuracy_score(y_test_dom[~np.isnan(y_pred_dom)], y_pred_dom[~np.isnan(y_pred_dom)]),
    "Precision": precision_score(y_test_dom[~np.isnan(y_pred_dom)], y_pred_dom[~np.isnan(y_pred_dom)]),
    "Recall": recall_score(y_test_dom[~np.isnan(y_pred_dom)], y_pred_dom[~np.isnan(y_pred_dom)]),
    "F1": f1_score(y_test_dom[~np.isnan(y_pred_dom)], y_pred_dom[~np.isnan(y_pred_dom)]),
    "ROC AUC": roc_auc_score(y_test_dom[~np.isnan(y_proba_dom)], y_proba_dom[~np.isnan(y_proba_dom)]),
    "MCC": matthews_corrcoef(y_test_dom[~np.isnan(y_pred_dom)], y_pred_dom[~np.isnan(y_pred_dom)])
}
print("=== IN-DOMAIN TEST METRICS ===")
print(pd.DataFrame(metrics, index=["Value"]).T.round(3))

# 11) Y-randomization (in-domain)
rand = {m: [] for m in metrics}
idx = ~np.isnan(y_pred_dom)
for _ in range(50):
    ys = np.random.permutation(y_test_dom[idx])
    pr = best.named_steps["clf"].predict(X_test_trans[idx])
    pb = best.named_steps["clf"].predict_proba(X_test_trans[idx])[:,1]
    rand["Accuracy"].append(accuracy_score(ys, pr))
    rand["Precision"].append(precision_score(ys, pr, zero_division=0))
    rand["Recall"].append(recall_score(ys, pr, zero_division=0))
    rand["F1"].append(f1_score(ys, pr, zero_division=0))
    rand["ROC AUC"].append(roc_auc_score(ys, pb))
    rand["MCC"].append(matthews_corrcoef(ys, pr))
rsum = []
for m,v in rand.items():
    mu, sd = np.mean(v), np.std(v,ddof=1)
    z = (observed - rand_mean) / rand_std if rand_std > 0 else np.nan
    p = 1 - ndtr(z) if not np.isnan(z) else np.nan
    rsum.append({"Metric":m, "Real":metrics[m], "RandMean±SD":f"{mu:.3f}±{sd:.3f}", "p-value":p})
print("\n=== Y-RANDOMIZATION ===")
print(pd.DataFrame(rsum).set_index("Metric"))

# ─── 11+12) RF & SHAP Feature Importances Side-by-Side ─────

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import shap

# 11) RF importances (post-corr filter)
keep_idx    = best.named_steps["corr"].keep
final_feats = np.array(feat_names)[keep_idx]  # Only the features kept after correlation filter
rf_imp      = best.named_steps["clf"].feature_importances_
# Ensure rf_imp and final_feats have the same length
assert len(rf_imp) == len(final_feats), f"rf_imp ({len(rf_imp)}) and final_feats ({len(final_feats)}) length mismatch"
rf_series   = pd.Series(rf_imp, index=final_feats).sort_values(ascending=False)

# 12) SHAP importances
#  a) Transform TEST set through the pipeline up to the classifier
X_imp = best.named_steps["imp"].transform(X_test)
X_sc  = best.named_steps["scale"].transform(X_imp)
X_cf  = best.named_steps["corr"].transform(X_sc)

#  b) Compute SHAP values
explainer = shap.TreeExplainer(best.named_steps["clf"])
shap_vals = explainer.shap_values(X_cf)

#  c) Extract the “positive‐class” SHAP array in a robust way
# For binary classification, shap_vals is a list of 2 arrays (n_samples, n_features)
if isinstance(shap_vals, list):
    arr = shap_vals[1]  # positive class
else:
    arr = shap_vals

# Ensure arr shape matches (n_samples, n_features)
if arr.shape[1] != len(final_feats):
    raise ValueError(f"SHAP values shape {arr.shape} does not match number of features {len(final_feats)}")

#  d) Mean(|SHAP|) across samples
# If shap_imp is 2D (n_features, 2), reduce to 1D by averaging over the last axis
shap_imp = np.abs(arr).mean(axis=0)
if shap_imp.ndim == 2:
    shap_imp = shap_imp.mean(axis=1)
shap_series = pd.Series(shap_imp, index=final_feats).sort_values(ascending=False)

# Plot them side by side
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

rf_series.head(20).plot(
    kind="barh", ax=axes[0], color="C0"
)
axes[0].invert_yaxis()
axes[0].set_title("Top 20 RF Feature Importances")
axes[0].set_xlabel("Gini Importance")

shap_series.head(20).plot(
    kind="barh", ax=axes[1], color="C1"
)
axes[1].invert_yaxis()
axes[1].set_title("Top 20 SHAP Mean(|Value|)")
axes[1].set_xlabel("Mean |SHAP value|")

plt.tight_layout()
plt.show()


# 14) Confusion matrix & ROC curve
cm = confusion_matrix(y_test_dom[idx], y_pred_dom[idx])
plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False)
plt.title("In-Domain Confusion Matrix"); plt.xlabel("Pred"); plt.ylabel("True")
plt.show()

fpr, tpr, _ = roc_curve(y_test_dom[idx], y_proba_dom[idx])
plt.figure(figsize=(6,4))
plt.plot(fpr, tpr, label=f"AUC={metrics['ROC AUC']:.3f}")
plt.plot([0,1],[0,1],'--',c='gray')
plt.title("In-Domain ROC Curve"); plt.xlabel("FPR"); plt.ylabel("TPR"); plt.legend()
plt.show()

print("\n=== WORKFLOW COMPLETE ===")
```

```python
# ── Applicability Domain on Training & Test Sets ──

def fit_applicability_domain(X_train, n_components=10, alpha=0.975, random_state=42):
    """
    Fit scaling + PCA on training data and determine Mahalanobis threshold.
    Returns scaler, pca, inverse covariance, mean, and chi2 threshold.
    """
    # Scale to zero mean/unit variance
    scaler = StandardScaler().fit(X_train)
    Xs = scaler.transform(X_train)
    
    # PCA reduction
    pca = PCA(n_components=n_components, random_state=random_state).fit(Xs)
    Xp = pca.transform(Xs)
    
    # Covariance in PCA space
    cov = np.cov(Xp, rowvar=False)
    inv_cov = np.linalg.inv(cov)
    mu = Xp.mean(axis=0)
    
    # Chi-squared threshold for given confidence level
    thresh = chi2.ppf(alpha, df=n_components)
    
    return scaler, pca, inv_cov, mu, thresh

def compute_applicability(X, scaler, pca, inv_cov, mu, thresh):
    """
    Compute Mahalanobis distance and in-domain flag for each sample in X.
    Returns distances and boolean in-domain mask.
    """
    Xs = scaler.transform(X)
    Xp = pca.transform(Xs)
    
    dists = []
    for vec in Xp:
        delta = vec - mu
        d2 = float(delta.dot(inv_cov).dot(delta))
        dists.append(d2)
    dists = np.array(dists)
    
    return dists, dists <= thresh

# Assuming X_train, X_test are already defined descriptor DataFrames

# 1) Fit AD on training set
scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh = fit_applicability_domain(
    X_train, n_components=10, alpha=0.975, random_state=42
)

print(f"Applicability Domain χ² threshold (97.5%): {chi2_thresh:.2f}")

# 2) Compute AD metrics on training set
train_dists, train_in_ad = compute_applicability(
    X_train, scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh
)
ad_train = pd.DataFrame({
    "Mahalanobis_D²": train_dists,
    "In_AD": train_in_ad
}, index=X_train.index)

# 3) Compute AD metrics on test set
test_dists, test_in_ad = compute_applicability(
    X_test, scaler_ad, pca_ad, inv_cov_ad, mu_ad, chi2_thresh
)
ad_test = pd.DataFrame({
    "Mahalanobis_D²": test_dists,
    "In_AD": test_in_ad
}, index=X_test.index)

# Display summary
print("\nTraining set AD coverage:")
print(ad_train["In_AD"].value_counts(), "\n")
print("Test set AD coverage:")
print(ad_test["In_AD"].value_counts())

# Optionally, append to your existing results tables:
# results_train = pd.concat([results_train, ad_train], axis=1)
# results_test  = pd.concat([results_test,  ad_test ], axis=1)
```
